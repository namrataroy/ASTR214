## Non linear regime

Many objects in the Universe (galaxies, clusters) have densities that are much higher than the mean density of the universe. These objects are in the non linear regime and the formation of such structure cannot be described by linear theory.

$$P(k,t)=P_i(k)T^2(k)D^2(t)$$

In the linear regime, all modes $k$ evolve similarly and and independently $\delta_k \propto D(a)$

## Top Hat Spherical Collapse

We now consider the highy idealized case of the *Top Hat Spherical Collapse Model*

Imagine the case of a completely smooth universe, except for a single, spherical, "top-hat" overdensity.

Note: although we call this a "top-hat" think more along the lines of a spherical water ballon floating in space .. See Figure 11.3 in Ryden.

Given the symmetry, the collapse of a spherical perturbation is a one-dimensional
problem and is fully specified by evolution of the tophat radius $R(t)$ (Gunn & Gott 1972, Lahav et al. 1991).

<img width=800 align=left src="slidesfigs/top_hat.png"></img>

Consider a spherically-symmetric density fluctuation:

* of initial radius $R_{\rm i}$
* amplitude $\delta_{i}$
* Mass $M=(4\pi/3)(1+\delta_{\rm i})\bar{\rho_m}R_{i}^3$
* $\bar{\rho}_{\rm m}$ is the mean *matter* density of the Universe at time t

We are going to write down the evolution of this spherical overdesnity of dark matter. We can model the overdensity as a series of many individual thin mass shells.

<img width=800 align=left src="slidesfigs/onion_model.png"></img>

How does the spherical overdensity evolve?
* First expands with the expansion of the universe. $R(t)$ increases as $R(t)\approx a(t)$ (following the "Hubble flow") as the amplitude of the perturbation is small and it expands almost at the expansion rate of the Universe.
* A given shell becomes bound and will begin to collapse. The time at which the shell reaches the maximum radius is called the *turn around radius*. The shell reaches this point at the *turn around time*.
* The shell collapses back down to $r=0$. This is called the *virialization time* or the *collapse time*
* Shells begin to oscillate and *shell crossing* occurs. At this point, shells begin to exchange energy and *virialization* occurs, resulting in the formation of a dark matter halo.

<img width=800 align=left src="slidesfigs/turn_around.png"></img>

Let's compute what happens *up to the point of shell crossing*.

We use Newton's second law of motion together with the shell theorem (the smooth distribution around the sphere exerts no force on that sphere)

$$\frac{d^2r}{dt^2}=-\frac{GM}{r^2}$$

Integrate this once. The integration constant is the specific energy of the shell $E$ (multiplying by m gives the conservation of energy equation $E_K+E_p=E_{tot}$). Then write that a gravitationally bound object has $E<0$

<img width=800 align=left src="slidesfigs/astro610_lecture8-09.png"></img>

<img width=800 align=left src="slidesfigs/astro610_lecture8-10.png"></img>

TO REMEMBER: $t_{vir}=2\times t_{ra}$

Now let's use *Energy conservation* to further solve the equation governing the collapse of the mass shell (this will tell us what A and B are)

For simplicity, let's assume an EDS universe ($\Omega_m=1$). Then:
* $\overline{\rho}=1/(6\pi G t^2)$

<img width=800 align=left src="slidesfigs/astro610_lecture8-11.png"></img>

REMEMBER: 
* For $\Omega_{\rm m}=\bar{\rho}_{\rm m}/\rho_{\rm crit}=1$, any overdensity $\delta_i>0$ will collapse. 
* For $\Omega_{\rm m}<1$, however, only perturbations with $\delta_{i}>1/\Omega_{\rm m}(t_i)-1$ will collapse.

## Virialization

The simple Spherical Collapse model is only valid up to the point of shell crossing (remember that at this point, $M(r)$ is no longer conserved). Instead, individual shells begin to interact gravitationally and to exchange energy, and virialization occurs.

Virialized means a system of gravitationaly interacting particles that is stable. The small structures interact with each other, but the cluster doesn't expand or collapse. A virialized object obeys the virial theorem which states that

$$2E_K + E_p=0$$

where $E_K$ is the kinetic energy and $E_p$ is the potential energy.

## Virial Overdensity

$$1+\Delta_{vir}=18\pi^2 \sim 178$$ 

<img width=800 align=left src="slidesfigs/astro610_lecture8-21.png"></img>

The reference for the values of $\Delta_{vir}$ in a non EDS cosmology is <a href="http://adsabs.harvard.edu/abs/1998ApJ...495...80B">Bryan and Norman 1998</a>.

$\Delta_{vir}$ form the basis of the definition of the "Virial Mass" of dark matter halos.

The virial mass is the mass enclosed by the radius such that the density interior is $\Delta_{vir} \times \overline{\rho}$

$$M_{vir}=(\frac{4}{3}\pi R_{vir}^3) \times \Delta_{vir} \times \overline{\rho}$$


**Note:** the mean density is sometimes written as either $\overline{\rho}$, $\rho_m$, $\rho_b$.

**Note:** in the litterature, you will see a variety of different definitions of "halo mass" and "halo radius". M$_{vir}$, M$_{200c}$, M$_{200m}$, M$_{500c}$. For example, M$_{200c}$ is defined as:

$$M_{200c}=(\frac{4}{3}\pi R_{vir}^3) \times 200 \times \rho_{\rm crit}$$

## Smoothing the density field with window functions

We will start by discussing how to smooth the density field - the utility of this exerciece will only become more apparent later on. 

<img width=800 align=left src="slidesfigs/astro610_lecture9-07.png"></img>

The normalization here is designed so that you don't actually modify on average the normalization of the density field. Hence, the integral over the window needs to be 1.

<img width=800 align=left src="slidesfigs/astro610_lecture9-08.png"></img>

Let's see what smoothing does to a density field ....

```python
# Here we create a "box" of dimension N
# We fill the box with a density field by drawing from a gaussian
# We normalize the density field by the expeced value for some cosmology and redshift
# We then compare the power spectrum in our box to our cosmological one!
# Beware: this takes some time to run! It will need to run before you can run the next step.

import numpy as np
from numpy.fft import fftn, ifftn, fftfreq
from numpy.random import normal
from matplotlib import pylab as plt

from colossus.cosmology import cosmology

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}

# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)
cosmo.interpolation = False

# Make a box with this size in h^{-1} Mpc
Lbox = 100.0

# grid size (change N to 128 if 256 is taking too long)
N=256
N3 = N*N*N
dx = Lbox/N
dxi = 1.0/dx

# set up grid coordinates

# N data points
x1d = np.linspace(0, N-1, N)    

# meshgrid(x,y,z) returns 3-D grid coordinates defined by the vectors x, y, and z
x, y, z = np.meshgrid(x1d, x1d, x1d)

# set random seed; remove if you want new realization every time
np.random.seed(141592653)

# generate a 3D array on our grid of Gaussian numbers with zero mean and unit dispersion
# Go back and look at at our class on Gaussian processes
# The value of delta at a random point is drawn from a gaussian distribution
delta = np.reshape(normal(0.0, 1.0, N3),(N,N,N))

# Fourier transform the delta field
# Read https://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.fftn.html
# What is the difference between fft and fftn?
deltak = fftn(delta)

# get vector of wavenumbers corresponding to modes
k1d = fftfreq(N) * 2.0 * np.pi

# grid of k in Fourier space
kx, ky, kz = np.meshgrid(k1d, k1d, k1d)

# Modulus of wavevector
k = dxi * np.sqrt(kx**2 + ky**2 + kz**2)
k[0,0,0] = np.pi * 2.0 / Lbox

# set up power spectrum for the input model at redshift zini
zini = 30.0

# growth factor at z=30 
gzi = cosmo.growthFactor(30.0)

# Remember that the redshift evolution of P(k) goes like D^2
gzi2 = gzi * gzi

# compute the power spectrum at z=30
Pk  = cosmo.matterPowerSpectrum(k, 'eh98') * gzi2 

# normalize Fourier components using sqrt(P(k))
# Remember how P(k) is defined .....
deltak *= np.sqrt(Pk)

# this is a "DC" or "fundamental mode" corresponding to the constant density in the box. 
# it can be set to any constant
deltak[0,0,0] = 0.0

# Now let's compute P(k) from the box that we created
lk = np.log10(k)
Nb = 20 # number of k bins to use

# P(k) is a weighted histogram of the squares of delta_k amplitudes
Pkj,  lkj = np.histogram(lk, bins=Nb, weights=np.square(np.absolute(deltak)), density=False) 

# <delta^4> for variance of the P(k)
Pkj2, lkj = np.histogram(lk, bins=Nb, weights=np.square(np.square(np.absolute(deltak))), density=False) 
lkaj, lkj = np.histogram(lk, bins=Nb, weights=lk, density=False) 
nj , lkj  = np.histogram(lk, bins=Nb, density=False) 

# Rms error on P(k) due to a finite number of modes
Pkj = Pkj / nj / gzi2 / N3
Pkj2 = Pkj2 / (gzi2*gzi2) / N3 / N3
sPkj = np.sqrt(Pkj2/(nj*(nj-1)) - Pkj**2/(nj-1))
lkaj = lkaj / nj 
kaj = 10.**lkaj

# transform the field back into real space
delta = np.real(ifftn(deltak))

# Let's get the input model spectrum for comparison
lka = np.arange(-4.,3.,0.02)
ka = 10.0**lka
Pka = cosmo.matterPowerSpectrum(ka, 'eh98')

#
# plot P(k) that we created and compare to the expected value
#
Pkjp = np.clip(Pkj,1.e-10,1.e100)
plt.figure(figsize=(3,3))
plt.xscale('log'); plt.yscale('log')
plt.ylim(1.e-2, 1.e5); plt.xlim(1.e-4, 1.e2)
plt.errorbar(kaj, Pkjp, yerr=2.*sPkj, fmt='o', markersize=1.)
plt.scatter(kaj, np.clip(Pkj,1.e-10, 1.e100), s=10., label=r'$\mathrm{realization}$')
plt.plot(ka, Pka, c='slateblue', linewidth=2., label=r'$\mathrm{input\ model}$')
plt.legend(frameon=False, loc='lower left', fontsize=10, scatterpoints=1)
plt.xlabel(r'$k\ (h\rm\ Mpc^{-1})$')
plt.ylabel(r'$P(k)\ (h^{-3}\ \rm Mpc^3)$')
plt.savefig('fig/pk_realization.png', bbox_inches='tight')
plt.show()
```

```python
# Now we take our density field, and we smooth it with a top hat filter
# Wfk is the filter

from scipy import signal

# Filter in Fourier Space
def Wfk(kR, filter = 'tophat'):
    """
    Top-hat and Gaussian window functions in Fourier space
    """
    if filter == 'tophat':
        return 3.*(np.sin(kR)-kR*np.cos(kR))/(kR)**3
    elif filter == 'gaussian':
        return np.exp(-0.5*kR**2)
    else:
        print("error in Wfk: routine accepts only tophat or gaussian filter")
        return

#
# plot a slice of the Gaussian density field
#
import matplotlib.ticker as mtick
plot_pretty()
lloc = 'upper right'; lfs = 9; lcol = 'white'; ccol = 'yellow'
fig, ax = plt.subplots(2,2,figsize=(4,4))


for ix in [0,1]:
    for iy in [0,1]:
        ax[ix][iy].set_xlim(0., Lbox); ax[ix][iy].set_ylim(0., Lbox)

dmin = 0.75*np.min(delta[:,:,N//2]); dmax = 0.5*np.max(delta[:,:,N//2])
x1d = Lbox*np.linspace(0, N-1, N)/(N-1)
x, y = np.meshgrid(x1d, x1d)

ax[0][0].imshow(delta[:,:,N//2], cmap='winter', vmin=dmin, vmax=dmax, extent=(0., Lbox, Lbox, 0.))
ax[0][0].plot(0, 0, "-", c='black', label=r'$\mathrm{no\ smoothing}$' )
#ax[0][0].contour(x, y, delta[:,:,N/2]/gzi.T, extent=(0., Lbox, 0., Lbox), linewidths=(0.5), colors=ccol, levels = [1.69])

#ax[0][0].colorbar(fraction=0.04, pad=0.04)
lgnd = ax[0][0].legend(frameon=False, loc=lloc, fontsize=lfs)
for text in lgnd.get_texts():
    text.set_color(lcol);
for item in lgnd.legendHandles:
    item.set_visible(False)
 
# smooth the field with a given filter size and plot
Rfa = [0.5, 1.5, 5.0] #/h Mpc
ix = [0, 1, 1]; iy = [1, 0, 1]
# smooth by a real-space top-hat filter
filter = 'tophat'

for ir, Rf in enumerate(Rfa):
    kR = k*Rf; 
    # A convolution is a multiplication in fourier space
    deltas = fftn(delta) * Wfk(kR, filter)
    deltas = np.real(ifftn(deltas))
    #ax[ix[ir]][iy[ir]].imshow(deltas[:,:,N/2], cmap='winter', vmin=dmin, vmax=dmax, extent=(0., Lbox, 0., Lbox) )
    ax[ix[ir]][iy[ir]].imshow(deltas[:,:,N//2], vmin=dmin, vmax=dmax, 
                              cmap='winter',  extent=(0., Lbox, Lbox, 0.) )
    delta_s = deltas[:,:,N//2]/gzi
    ax[ix[ir]][iy[ir]].contour(x, y, delta_s,  linewidths=(0.5), colors=ccol, 
                               levels = [1.69])
    ax[ix[ir]][iy[ir]].plot(0, 0, "-", c='black', label=r'$R_{\rm f}=%.1fh^{-1}\ \mathrm{Mpc}$'%Rf )
    lgnd = ax[ix[ir]][iy[ir]].legend(frameon=False, loc=lloc, fontsize=lfs)
    for text in lgnd.get_texts():
        text.set_color(lcol);
    for item in lgnd.legendHandles:
        item.set_visible(False)

plt.setp(ax[0][0].get_xticklabels(), visible=False)
plt.setp(ax[0][1].get_xticklabels(), visible=False)
plt.setp(ax[0][1].get_yticklabels(), visible=False)
plt.setp(ax[1][1].get_yticklabels(), visible=False)
ax[0][0].set_ylabel(r'$y\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][0].set_ylabel(r'$y\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][0].set_xlabel(r'$x\ (h^{-1}\ \mathrm{Mpc})$')
ax[1][1].set_xlabel(r'$x\ (h^{-1}\ \mathrm{Mpc})$')

plt.tight_layout()

plt.savefig('fig/gfield_smooth.png', bbox_inches='tight')
plt.show()
```

### Variance of the smoothed density field

For the cases, when one is interested in only a narrow range of $k$ the power spectrum can be
approximated by a power-law form, $P(k)\propto k^n$, and the
variance is $\sigma^2(R)\propto R^{-(n+3)}$ where n is the local spectral index at that perturbation wavelength.

The characteristic amplitude of peaks in the $\delta_R$ (or $\delta_{M}$)
field smoothed on scale $R$ (or mass scale $M$) is given by
$\sigma(R)\equiv\sigma(M)$.  

The smoothed Gaussian density field is also Gaussian with the probability distribution function
(PDF) given by 
$$
p(\delta_{M})=\frac{1}{\sqrt{2\pi}\sigma(M)}\exp\left[-\frac{\delta^2_{M}}{2\sigma^2(M)}\right].
$$

## Peak Height and Collapsed Regions

We are now going to connected two different things we have learned so far: $\delta_c$ and $\sigma(M)$ and will learn about a third concept: peak height.

Recall that $\delta_c$ tells us regions in the linearly extrapolated density field with $\delta>\delta_c$ will collapse.

Let's look at the redshift evolution. Regions are growing with time and $\delta$ evolves like $D(t)$. We will simply move $D(t)$ to the other side of the equation to view $\delta$ as fixed by $\delta_c$ evolving with time.

The collapse barrier moves downward with time - smaller things collapse first, and larger things collapse later.

<img width=800 align=left src="slidesfigs/astro610_lecture9-05.png"></img>

<img width=800 align=left src="slidesfigs/astro610_lecture9-04.png"></img>

The distribution of peak amplitudes in the initial Gaussian overdensity field smoothed over radius (or mass) scale R (or M) is given by a Gaussian PDF with a rms value of $\sigma(M)$. 

The peaks in the initial Gaussian overdensity field smoothed at redshift $zi$ over mass scale M can be characterized by the ratio $\nu = \delta_c / \sigma(M, zi)$ called the **peak height**. 

The typical peaks which are collapsing at any given redshift are those for which $\nu=1$

## Press Schecter Mass Function (in a nutshell)

In their pioneering model, <a href="http://adsabs.harvard.edu/abs/1974ApJ...187..425P">Press & Schechter (1974)</a> adopted the ansatz motivated by the spherical top-hat collapse model  that any point in space with $\delta_{\rm M}({\bf x})D_{+0}(z)\geq \delta_c\approx 1.69\approx\mathrm{const}$ will collapse into a halo of mass $\geq M$ by redshift $z$

Following this assumption, they showed that **the abundance of halos of mass $M$ at redshift $z$ is a universal function of only their peak height** $\nu(M,z)\equiv\delta_c/\sigma(M,z).$

In the Press-Schechter ansatz (<a href="http://adsabs.harvard.edu/abs/1974ApJ...187..425P">Press & Schechter 1974</a>) the mass function of halos has the form

$$\frac{dn_{\rm PS}}{d\ln M}=\sqrt{\frac{2}{\pi}}\frac{\bar{\rho}}{M}\nu\frac{d\ln\nu}{d\ln M} \exp\left(-\frac{\nu^2}{2}\right)$$

* n is the number density of dark matter halos per unit volume
* $\nu$ is the peak height
* M is halo mass

Note, however, that this is only an approximation, and although it matches predictions from N-body simulations, it is not precise enough for most cosmology research today.

Recall that $\nu\propto 1/\sigma(M)\propto M^{-\alpha}$, where $\alpha$ depends on the slope of the power spectrum which changes with scale. 



**Recall**: the power spectrum slope depends on wavenumber $k=2\pi/\lambda$, where $\lambda=2R$ is the wavelength of the perturbation. 



The above form translates into the following mass function as a function of halo mass $M$: 
* approximately a power law at $M\ll M_*$
* exponential cutoff at $M\gg M_*$ (notice the exponential part in the Press Shechter halo mass function). 

```python
# Plot the PS Halo Mass Function

import numpy as np
from matplotlib import pylab as plt
from colossus.cosmology import cosmology
import matplotlib.cm as cm
import matplotlib.colors as colors

# define a vector of cosmological parameters:    
my_cosmo = {'flat': True, 'H0': 70.0, 'Om0': 0.27, 'Ob0': 0.045714, 'sigma8': 0.82, 'ns': 0.96}

# set my_cosmo to be the current cosmology	
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo)

# grid of halo masses in /h Msun
lmh = np.arange(6.,15.5,0.05)
mh = np.power(10.,lmh)

# compute the Lagrangian radius corresponding to perturbation of mass mh in /h Mpc
# This converts the mass of a halo (in comoving Msun/h to the radius of its comoving Lagrangian volume
# (in comoving Mpc/h), that is the volume that encloses the halos mass given the mean density
# of the universe at z = 0.
# simply does: return (3.0 * M / 4.0 / np.pi / self.rho_m(0.0) / 1E9)**(1.0 / 3.0)
rh = cosmo.lagrangianR(mh)

# compute sigma(R) using default Eisenstein & Hu 1998 power spectrum and tophat filter
sig = cosmo.sigma(rh,z=0.)

# and its logarithmic derivative dlog(sigma)/dlogR
dsdr = cosmo.sigma(rh, z=0., derivative='True')

# dln(sigma)/dlnR -> dln(sigma)/dlnM, M = (4*pi/3) rho_mean * R^3
dsdm = dsdr / 3.0

#  Press-Schechter mass function parameters
fig = plt.figure(figsize = (3,3))
plt.xlabel(r'$M\ (h^{-1} M_{\odot})$'); plt.ylabel(r'$dn/d\ln M\ (h^3\ \rm Mpc^{-3})$')
plt.yscale('log'); plt.xscale('log')
plt.xlim(1.e6, 4.e15); plt.ylim(1.e-7,1.e3)
nu = cosmo.peakHeight(mh, 0.0)

# The equation we wrote above
nps = np.sqrt(2.0/np.pi)*cosmo.Om0*2.77e11/mh*nu*np.abs(dsdm)*np.exp(-0.5*nu**2)

plt.plot(mh,nps,c='darkslateblue', linewidth=2.0, label=r'$z=0$')
    
plt.legend(frameon=False, loc='upper right', fontsize=9)
plt.grid(linestyle='dotted', c='lightgray')
plt.savefig('fig/mf_ps.png', bbox_inches='tight')
plt.show()
```

There are other more recent analytic ways to compute the Halo Mass Function. Modern ones are *calibrated* to match the results from N-body simulations.

Developping more accurate calibrations of the HMF is a field of active research. The level of accuracy of calibrations of the HMF are are the 10% level.

Historically, the problem with such calibrations has been that calbrations have been limited by the availability of only a small number of N-body simulations with a handfull of cosmologies.

Because of our expanding ability to compute many N-body simulations over a range of cosmologies, there are now groups that are using *emulator* (a form of interpolation) techniques to build more accurate HMFs.
